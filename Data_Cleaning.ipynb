```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("melb_data.csv")
print(data.shape)
data.head()
```

```python
data['Unique ID'] = data['Unique ID'].apply(lambda x:int(x[3:]))
data.set_index("Unique ID", inplace=True)
data.tail()
```

```python
cat_features = [feature for feature in data.columns if data[feature].dtypes == 'O']
cat_features
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data[cat_features] = data[cat_features].apply(lambda x:le.fit_transform(x))
print(data.dtypes)
```

```python
features_with_na=[features for features in data.columns if data[features].isnull().sum()>1]
data[features_with_na] = data[features_with_na].apply(lambda x:x.fillna(x.median()))
data.info()
```

```python
data['Car'] = data['Car'].astype(int)
data['YearBuilt'] = data['YearBuilt'].astype(int)
```

```python
data.describe()
```

```python
print("Mean value of Rooms feature :", data["Rooms"].mean())
print("Median value of Rooms feature :", data["Rooms"].median())
print("Mode of Rooms feature :", data["Rooms"].mode()[0])
print("Variance of Rooms feature :", data["Rooms"].var(),"\n")

print("Mean value of Price feature :", data["Price"].mean())
print("Median value of Price feature :", data["Price"].median())
print("Mode of Price feature :", data["Price"].mode()[0])
print("Variance of Price feature :", data["Price"].var())
```
